```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
import pandas as pd
import numpy as np
df=pd.read_csv('ausdata.csv')
# Define the time series data
for i in range(10):
  # Define the time series data
    data = list(df['Value'][i*11:i*11+10])
    drug=df['Variable'][i*11+10]
# Define the number of time steps for the LSTM model
    n_steps = 3

# Split the data into input (X) and output (y) sequences
    X, y = [], []
    for i in range(len(data)-n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])

# Reshape the input data to be 3D for LSTM
    X = np.array(X).reshape(len(X), n_steps, 1)

# Define the LSTM model
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    y=np.array(y)
# Fit the LSTM model to the data
    model.fit(X, y, epochs=200, verbose=0)

# Use the LSTM model to predict the next value in the time series
    last_sequence = np.array(data[-n_steps:]).reshape(1, n_steps, 1)
    prediction = model.predict(last_sequence)

    print("Next value in the time series: ", prediction[0][0],drug)
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('ausdata.csv')

# Define the number of time steps for the LSTM model
n_steps = 3

# Loop through the first 10 rows of data in the DataFrame
for i in range(10):
    # Define the time series data and drug name for the current row
    data = list(df['Value'][i*11:i*11+10])
    drug=df['Variable'][i*11+10]

    # Split the data into input (X) and output (y) sequences
    X, y = [], []
    for i in range(len(data)-n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])

    # Reshape the input data to be 3D for LSTM
    X = np.array(X).reshape(len(X), n_steps, 1)

    # Define the LSTM model
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')

    # Convert the output data y to a numpy array
    y=np.array(y)

    # Fit the LSTM model to the data
    model.fit(X, y, epochs=200, verbose=0)

    # Use the LSTM model to predict the next value in the time series
    last_sequence = np.array(data[-n_steps:]).reshape(1, n_steps, 1)
    prediction = model.predict(last_sequence)

    # Append the predicted value to the original time series data
    data.append(prediction[0][0])

    # Plot the original and predicted time series data
    plt.plot(data, label='Original')
    plt.plot(range(len(data)-1, len(data)+1), [data[-2], prediction[0][0]], marker='o', color='red', label='Predicted')
    plt.title(drug)
    plt.legend()
    plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ausdata.csv')

for i in range(10):
    data = list(df['Value'][i*11:i*11+10])
    model = ARIMA(data, order=(1, 1, 1)) # This specifies an ARIMA(1,1,1) model
    model_fit = model.fit()

    # Predict the next value in the time series
    forecast = model_fit.forecast()
    drug = df['Variable'][i*11+10]

    print("Next value in the time series: ", forecast, drug)

    # Plot actual vs predicted values
    plt.plot(data, label='Actual')
    plt.plot([10], forecast, marker='o', markersize=8, label='Predicted')
    plt.title(drug)
    plt.legend()
    plt.show()
```
